The **second dimension** of a data analysis problem is the **variable dimension.** The  the main algorithms for that dimension are the **regression** and **classification** algorithms (supervized learing in machine learning terms). As we saw, the variables one choose to consider for our target population allow us to carry a "population analysis" by giving us criteria (the variable values) to group together individuals with similar  values for the variable at hand. An orthogonal question to that grouping is the grouping of variables into groups entertaining a certain dependence (covariance) in the values that variables in the same group can take. 

Here we will mostly focus on **linear regression techniques** to analyse **(linear) variable relationships** between the variables.


Models have 3 components:
1) model parameters
2) Prediction
3) residual errors = observations - predictions
The rough idea of fitting: Find parameter values such that the difference between the observed and what your model predicts given the observed values of the explanatory variables, the residual error, is minimized.

y = beta * x + e
where beta is the parameter, x is the observed value and e is the error

### Simulating explanatory variables: X

```python
%%R

clean.range = function(X, var.range){
        X[X < var.range[1]] = var.range[1]
        X[X > var.range[2]] = var.range[2]
        return(X)
    }
```

```python
%%R

simulate.X = function(var.size=1, var.mean=0, var.sd=1, var.range=NULL)
{
    X = rnorm(var.size, mean=var.mean, sd=var.sd)
    if (!is.null(var.range)) X = clean.range(X, var.range)
    return(X)
}
```

```python
%%R -r 86 -w 300 -h 300
X = simulate.X(var.size=200, var.mean=40, var.sd=20, var.range=c(0,100))
hist(X)
```

### Simulating target variables: Y

```python
%%R

simulate.Y = function(X.variables,      #
                      alpha,            #
                      beta,             #
                      noise.sd=1,    #
                      noise.mean=0,  #
                      var.range=NULL    #
                     )
    {
        X = as.matrix(X.variables)
        noise = rnorm(dim(X)[1], mean=noise.mean, sd=noise.sd)
        e = matrix(noise, ncol=1)
        v = matrix(beta, ncol=1)
    
        Y = alpha + X %*% v + e
    
        if(!is.null(var.range)) Y = clean.range(Y, var.range)
        return(as.numeric(Y))
    }
```

```python
%%R -r 86 -w 300 -h 300

Xsize=1000; Xmean = 0; Xsd = 100;

alpha = 2
beta  = 1
noiseStrength = 20

X = simulate.X(Xsize, Xmean, Xsd)
Y = simulate.Y(X, alpha, beta, noiseStrength)
```

```python
%%R -r 86 -w 400 -h 400

plot(X,Y)
```

### Simulating clusters

```python
%%R
#cluster informations
sizes.default   = c(30, 40, 50)
means.default   = c(-50, 100, 50)
sd.default      = c(10, 10, 10)
alphas.default  = c(20, 10, 30)
betas.default   = c(0.9, 1.2, -0.9)
errors.default  = c(10, 0, 20)
```

```python
%%R

clusters.info = function(cl.sizes=sizes.default,     #
                         cl.means=means.default,     #
                         cl.sd=sd.default,           #
                         cl.alphas=alphas.default,   #
                         cl.betas=betas.default,     #
                         cl.errors=errors.default    #
                        )
    {
        cl.list     = list()
        cl.number   = length(cl.sizes)
        sample.size = sum(cl.sizes)
        
        for (i in 1:cl.number){
            cl.name = paste('cluster', i, sep='')
            cl.list[[cl.name]] = c(size=cl.sizes[i],    #
                                   mean=cl.means[i],    #
                                   sd=cl.sd[i],         #
                                   alpha=cl.alphas[i],  #
                                   beta=cl.betas[i],    #
                                   error=cl.errors[i])  #
        }
        return(t(as.data.frame(cl.list)))
    }
```

```python
%%R

clusters.default = clusters.info()
print(clusters.default)
```

Alpha is the intercept, beta is the slope.

```python
%%R

simulate.bivariateCluster = function(cl.info){
    X = simulate.X(cl.info['size'], cl.info['mean'], cl.info['sd'])
    Y = simulate.Y(X, cl.info['alpha'], cl.info['beta'], cl.info['error'])
    return(data.frame(X=X,Y=Y))
}
    
```

```python
%%R

simulate.bivariateData = function(clusters.info=clusters.default)
{
    m = nrow(clusters.info) # the number of clusters
    data = data.frame()
    for(i in 1:m){
        df = simulate.bivariateCluster(clusters.info[i,])
        df$Cluster = factor(rep(i, nrow(df)))
        data = rbind(data, df)
    }
    shuffle = sample(nrow(data))# puts the rows (numbers) in random order in a vector
    data = data[shuffle, ] # Makes data frame with the rows in the shuffled order
    return(data)
}
```

## Computing and visualizing a regression

The difference between the predicted outome by the model and the observation is the residual for that data point. The sum of the squares of the residuals should be small. (since small is a relative term, it's not a reliable way to prove your model). the square root of the sum of squares over n - (m + 1) where m is the number of X's variables. is approximately equaly to the units of response (google this), otherwise known as the noice.

```python
%%R

model = lm(data$Y ~ data$X)
print(model)
```

```python
%%R

model = lm(data$Y ~ data$X)
print(summary(model))
```

We see that the residual standard error is 51, which is pretty high considering the max score is 100. The standard error should be low relative to the number of response.
If the residuals are normal, symmetrical about 0. (median = 0 and max-min = 0), then our model fits pretty well.
Above we see that our model doesn't fit very well, even though we have a linear relationship in the data.

It's always response - exaplanatory. Don't worry too much about the details of these statistics now, since it's a bit far outside the scope of the course.

```python
%%R -r 86 -w 300 -h 300

variable.number = 10
sample.size     = 200
variable.mean   = 100
variable.sd     = 200

X = replicate(variable.number, rnorm(sample.size, variable.mean, variable.sd))

colnames(X) = paste('X', 1:variable.number, sep='')

print(head(X))
```

```python
%%R -r 86 -w 600 -h 300

alpha = 10
beta  = rnorm(variable.number)
error = rep(0, variable.number)#rnorm(variable.number)

Y = simulate.Y(X, alpha, beta, error)

par(mfrow=c(1,2))
hist(Y)
boxplot(Y)
```

```python
%%R

df = data.frame(X, Y=Y)
#print(head(df))

ind = df$Y > 2000

df.outliers = df[ind,]
print(df.outliers)
```

```python
%%R -r 86 -w 1000 -h 1000

plot(df)
```

We want to uncover a linear relation of the form:

$$Y = \alpha + \beta_1 X_1 +\cdots + \beta_n X_n$$

That is: We want to find 

$\alpha$ and $\beta = (\beta_1,\dots,\beta_n)$ 

such that the error committed on our sample (here <code>df</code>) 

$$E_s = \|Y - (\alpha + X\beta)\|^2$$ 

is the smallest possible. The **residual** is the following vector computed on the sample:

$$\epsilon = Y - (\alpha + X\beta) = (\epsilon_1, \dots, \epsilon_m),$$

where the $i^{th}$ component indicates the error commited between the sample value of the $i^{ith}$ observation and the value predicted by the model for this observation: i.e.,

$$\epsilon_i = y_i - (\alpha - \beta_1 x_{i1}+\cdots + \beta_n x_{in}).$$

An **an assumption of our model is that the residual are normally distributed around zero.** It can be interpreted as a **random error** or a **noise** in our data.

Using the linear model <code>lm</code>, we would like to see the strength of the correlation:

$$Y = \alpha + X\beta + \epsilon.$$


This is the generalization to several variables of what we just saw. 

```python
%%R

model = lm(Y ~ ., data=df)
stats = summary(model)

print(stats)
```

We see some good values for the residuals. Very symmetric about 0. The standard error is very near zero.
